```python
# To load pickle files
import os, warnings, sys, pickle, gzip
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

sn.set(rc={"figure.dpi":200, 'savefig.dpi':200})
sn.set_context('notebook')
sn.set_style("ticks")
plt.rcParams["font.family"] = "Times New Roman"
cmap=sn.color_palette("viridis", as_cmap=True)
sn.set_theme()
```

# Loading data

Pick one of the following filenames:

### From section 3.1
* `data_SUSY_P11169.p`: SUSY data for $\mathbb{CP}[1,1,1,6,9]$
* `data_nonSUSY_P11169.p`: non-SUSY data for $\mathbb{CP}[1,1,1,6,9]$

### From section 3.2

* `data_SUSY_h12_2.p`: SUSY data for models at $h^{1,2}=2$
* `data_nonSUSY_h12_2.p`: non-SUSY data for models at $h^{1,2}=2$
* `data_SUSY_h12_3.p`: SUSY data for models at $h^{1,2}=3$
* `data_nonSUSY_h12_3.p`: non-SUSY data for models at $h^{1,2}=3$



```python
filename = os.getcwd()+"/data_SUSY_h12_2.p"

with gzip.open(filename, 'rb') as f:
    df = pickle.load(f)

f.close()
```

## Plotting moduli VEVs

```python
max_VEV=10

sn.scatterplot(x=df["Im(Z1)"],y=df["Im(Z2)"],s=5,alpha=1)
    
plt.rc('legend', fontsize=8)

plt.legend("",frameon=False)

plt.xlabel('Im$(Z_1)$')
plt.ylabel('Im$(Z_2)$')

plt.xlim(0,max_VEV)
plt.ylim(0,max_VEV)

plt.show();
```

## Plotting axio-dilaton VEV

```python
max_VEV=10

sn.scatterplot(x=df["Re(tau)"],y=df["Im(tau)"],s=5,alpha=1)
    
plt.rc('legend', fontsize=8)

plt.legend("",frameon=False)

plt.xlabel(r'Re$(\tau)$')
plt.ylabel(r'Im$(\tau)$')

plt.xlim(-0.6,0.6)
plt.ylim(0,max_VEV)

plt.show();
```
